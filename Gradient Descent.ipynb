```python
import numpy as np

def gradient_descent(x,y,lr=0.08,iterations = 10000, cost_threshold = 0.0012):
    m_curr = b_curr = 0
    iterations = iterations
    n = len(x)
    learning_rate = lr
    iteration = 1
    while  iteration<=iterations:
        y_predicted = m_curr * x + b_curr
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])
        dm = -(2/n)*sum(x*(y-y_predicted))
        db = -(2/n)*sum(y-y_predicted)
        print ("m {}, b {}, cost {} iteration {}".format(m_curr,b_curr,cost, iteration))
        #if cost <cost_threshold:
         #   break
        #Uncomment above to include cost and threshold    
        m_curr -= learning_rate * dm
        b_curr -= learning_rate * db
        iteration +=1


x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
gradient_descent(x,y,lr=0.08,iterations = 10000,cost_threshold = 0.0012)
```
